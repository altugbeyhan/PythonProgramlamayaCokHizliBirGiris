## Koşullu Durumlar

### if-elif-else

if: eğer

else: aksi halde

elif: else if (2'den fazla durum)

**KOŞULLAR**

eşittir: ==

eşit değildir: !=

küçüktür: <

büyüktür: >

küçük eşittir: <=

büyük eşittir: >=

```python
x = 10

if x == 10:
    print("sayı 10'a eşit")
```

```python
x = 20

if x == 10:
    print("sayı 10'a eşit")
```

```python
x = 20

if x == 10:
    print("sayı 10'a eşit")
else:
    print("sayı 10'a eşit değil")
```

```python
x = 7

if x != 5:
    print("sayı 5'e eşit değil")
else:
    print("sayı 5'e eşit")
```

```python
sayi = 13

if sayi < 12:
    print("sayı 12'den küçük")

elif sayi == 12:
    print("sayı 12'ye eşit")

else:
    print("sayı 12'den büyük")
```

## Döngüler

### While Döngüsü

while: -iken

while KOŞUL: KOŞUL sağlanıyor iken:

```python
x = 1

while x <= 12:
    print(f"{x}. sınıf") #format print
    x = x + 1 # x'i değiştirmezsek sonsuz döngü olur!
```
